####Este projeto tem como objetivo analisar o desempenho acadêmico de alunos com base em suas notas e frequência.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("notas_alunos.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# criando coluna para a média geral das notas

df['media_geral'] = (df['nota_prova1'] + df['nota_prova2']) / 2
```

```python
df.head(1)
```

```python
# filtro para descobrir médias acima de 9.00

df_nota_maior_9 = df[df['media_geral'] > 9.00]
df_nota_maior_9
```

```python
# aluno com maior frequencia

df_maior_frequencia = df[df['frequencia'] == df['frequencia'].max()]
df_maior_frequencia
```

```python
plt.scatter(df['media_geral'],df['frequencia'])
plt.title('Frequência x Média Geral')
plt.xlabel('Média Geral')
plt.ylabel('Frequência')
plt.show()
```

#### Há uma forte realção entre a frequência nas aulas com a média geral, pois o gráfico de dispersão mostra como os dados estão agrupados e seguem uma tendência.

```python
# descobrindo a maior e menor média

maior_media = df['media_geral'].max()
menor_media = df['media_geral'].min()

print(f'Maior média: {maior_media}')
print(f'Menor média: {menor_media}')
```

```python
# aluno com menor média

aluno_menor_media = df[df['media_geral'] == df['media_geral'].min()]
aluno_menor_media
```

```python
# aluno com menor frequência

aluno_menor_frequencia = df[df['frequencia'] == df['frequencia'].min()]
aluno_menor_frequencia
```

####Podemos afirmar que a frequência nas aulas implica para os desempenhos nas provas, onde o aluno com menor presença nas aulas teve a menor média da sala, por outro lado os estudantes com maior número de presenças nas aula obtiveram uma média próxima a 10.

```python
# gráfico para analisar como está a distribuição da média dos estudantes

plt.hist(df['media_geral'], bins=20, color='green',edgecolor='black')
plt.title('Distribuição da Média dos Alunos')
plt.xlabel('Média')
plt.ylabel('Frequência')
plt.show()
```

####O gráfico mostra que o desempenho da turma é predominantemente satisfatório, com a maior parte dos alunos concentrada acima da média mínima esperada.

```python
# criando uma coluna para determinar se o aluno foi Aprovado ou Reprovado

df['Aprovacao'] = df.apply(lambda row: 'Aprovado' if row['media_geral'] >= 6 and row['frequencia'] >= 75 else 'Reprovado', axis=1)
df.head()
```

####A análise dos dados indica que a maioria dos alunos foi aprovada, considerando a regra de negócio definida (média geral ≥ 6 e frequência ≥ 75%). A taxa de aprovação é elevada, o que sugere que o desempenho acadêmico geral da turma é satisfatório.

####A distribuição das médias mostra uma concentração predominante entre 6 e 8, reforçando que grande parte dos alunos apresenta desempenho consistente, com poucos casos de médias muito baixas. Além disso, a presença de algumas médias altas indica que há alunos com desempenho acima da média.

####Do ponto de vista analítico, não há indícios de um cenário preocupante, pois os resultados não demonstram concentração excessiva de reprovações nem grande dispersão negativa das notas. Assim, conclui-se que o desempenho geral da turma é positivo, com boa taxa de aprovação e distribuição equilibrada das médias.
